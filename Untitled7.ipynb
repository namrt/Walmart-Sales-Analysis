#Hypothesis test

```
November=df[df['Month'] == 12]['Sales(In ThousandDollars)']
```

```
NN=df[df['Month'] != 12]['Sales(In ThousandDollars)']
```

```


from scipy import stats
stats.ttest_ind(November.dropna(), NN.dropna())
#REJECT THE NULL HYPOTHESIS, NOVEMBER HAS MAX SALES
```
